```python
import random

def hangman():
    word_list = ["python", "programming", "computer", "algorithm", "developer", "challenge"]
    chosen_word = random.choice(word_list).upper()
    word_display = ["_" for _ in chosen_word]
    guessed_letters = set()
    lives = 6

    print("Welcome to Hangman!")
    print(f"The word has {len(chosen_word)} letters.")

    while lives > 0 and "_" in word_display:
        print("\n" + " ".join(word_display))
        print(f"Lives remaining: {lives}")
        print(f"Guessed letters: {', '.join(sorted(list(guessed_letters)))}")

        guess = input("Guess a letter: ").upper()

        if len(guess) != 1 or not guess.isalpha():
            print("Invalid input. Please enter a single letter.")
            continue

        if guess in guessed_letters:
            print("You already guessed that letter. Try again.")
            continue

        guessed_letters.add(guess)

        if guess in chosen_word:
            print(f"Good job! '{guess}' is in the word.")
            for i, letter in enumerate(chosen_word):
                if letter == guess:
                    word_display[i] = guess
        else:
            print(f"Sorry, '{guess}' is not in the word.")
            lives -= 1

    if "_" not in word_display:
        print("\n" + " ".join(word_display))
        print("Congratulations! You guessed the word and won!")
    else:
        print("\n" + " ".join(word_display))
        print("Game Over! You ran out of lives.")
        print(f"The word was: {chosen_word}")

if __name__ == "__main__":
    hangman()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

X=[[25,50000],[30,60000],[35,65000],[40,80000],[45,85000],[20,40000],[50,90000],[23,45000]]
y=[0,0,1,1,1,0,1,0]

m=DecisionTreeClassifier(criterion='entropy',max_depth=3,random_state=0).fit(X,y)
p=m.predict(X)

print("Accuracy:",accuracy_score(y,p))
print("\nConfusion Matrix:\n",confusion_matrix(y,p))
print("\nClassification Report:\n",classification_report(y,p))

print("Predictions:",m.predict([[28,55000],[42,78000],[22,41000]]))
```

```python
# Binary Classification using Decision Tree (custom data)

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Example custom dataset
# Features: [age, salary]
X = [
    [25, 50000],
    [30, 60000],
    [35, 65000],
    [40, 80000],
    [45, 85000],
    [20, 40000],
    [50, 90000],
    [23, 45000]
]

# Labels: 0 = "Not Buy", 1 = "Buy"
y = [0, 0, 1, 1, 1, 0, 1, 0]

# Create and train the Decision Tree model
model = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=0)
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Evaluate the model
print("Accuracy:", accuracy_score(y, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y, y_pred))
print("\nClassification Report:\n", classification_report(y, y_pred))

# Predict on new data
X_test = [[28, 55000], [42, 78000], [22, 41000]]
test_pred = model.predict(X_test)
print("Predictions for new data:", test_pred)
```
