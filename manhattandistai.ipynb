```python
def misplaced_tiles(current_state, goal_state):
    count = 0
    for i in range(3):
        for j in range(3):
            if current_state[i][j] != 0 and current_state[i][j] != goal_state[i][j]:
                count += 1
    return count


current = [
    [8, 7, 6],
    [5, 4, 3],
    [2, 1, 0]
]

goal = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 0]
]

print("Misplaced tiles:", misplaced_tiles(current, goal))
```

```python
def manhattan_distance(current_state, goal_state):
    distance = 0
    goal_positions = {}

    for i in range(3):
        for j in range(3):
            goal_positions[goal_state[i][j]] = (i, j)

    print("Tile-wise Manhattan distances:")
    for i in range(3):
        for j in range(3):
            value = current_state[i][j]
            if value != 0:
                goal_i, goal_j = goal_positions[value]
                tile_distance = abs(i - goal_i) + abs(j - goal_j)
                print(f"Tile {value}: Current=({i},{j}) Goal=({goal_i},{goal_j}) → Distance = {tile_distance}")
                distance += tile_distance

    print(f"\nTotal Manhattan distance: {distance}")
    return distance

current = [
    [8, 7, 6],
    [5, 4, 3],
    [2, 1, 0]
]

goal = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 0]
]

manhattan_distance(current, goal)
```
