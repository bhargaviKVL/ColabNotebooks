```python
# Valid positions and states
positions = ['window', 'middle', 'door']
vertical_positions = ['floor', 'box']

# Initial state
monkey_hpos = input("Monkey horizontal start (window/middle/door): ").strip().lower()
monkey_vpos = input("Monkey vertical start (floor/box): ").strip().lower()
box_pos = input("Box start position (window/middle/door): ").strip().lower()

def print_state():
    print(f"Monkey is at {monkey_hpos} ({monkey_vpos}), Box is at {box_pos}")

def move_monkey(to_pos):
    global monkey_hpos
    if to_pos in positions:
        print(f"Monkey moves from {monkey_hpos} to {to_pos} (on {monkey_vpos})")
        monkey_hpos = to_pos
        return True
    else:
        print(f"Invalid position to move: {to_pos}")
        return False

def push_box(to_pos):
    global monkey_hpos, box_pos, monkey_vpos
    if monkey_vpos != 'floor':
        print("Monkey must be on the floor to push the box.")
        return False
    if monkey_hpos != box_pos:
        print("Monkey must be at the box position to push it.")
        return False
    if to_pos not in positions:
        print(f"Invalid position to push box: {to_pos}")
        return False

    print(f"Monkey pushes the box from {box_pos} to {to_pos}")
    box_pos = to_pos
    monkey_hpos = to_pos
    return True

def climb_box():
    global monkey_vpos, monkey_hpos, box_pos
    if monkey_vpos == 'box':
        print("Monkey is already on the box.")
        return False
    if monkey_hpos != box_pos:
        print("Monkey must be at the box position to climb it.")
        return False
    print("Monkey climbs onto the box.")
    monkey_vpos = 'box'
    return True

def climb_down():
    global monkey_vpos
    if monkey_vpos == 'floor':
        print("Monkey is already on the floor.")
        return False
    print("Monkey climbs down from the box.")
    monkey_vpos = 'floor'
    return True

def grab_banana():
    if monkey_vpos == 'box' and monkey_hpos == 'middle' and box_pos == 'middle':
        print("Monkey grabs the banana!")
        return True
    print("Monkey cannot reach the banana.")
    return False


# Simulation of the solution:

print("\nInitial state:")
print_state()

# Move monkey to box if not there
if monkey_hpos != box_pos:
    move_monkey(box_pos)

# Push box to middle if not there
if box_pos != 'middle':
    push_box('middle')

# Climb on box if not already on it
if monkey_vpos != 'box':
    climb_box()

# Attempt to grab banana
success = grab_banana()

if success:
    print("\nSuccess! Monkey got the banana.")
else:
    print("\nFailure. Monkey couldn't get the banana.")
```
