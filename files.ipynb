WAP to find number of characters and words and lines in a file

```python
f = open('kvlb.txt','w')
f.write('hello\ngood morning\nmake some breakfast\n')
f.close()

f = open('kvlb.txt', 'r')
x = f.readlines()

print(x)

sum=0
for i in x:
  sum += len(i)

print('number of characters =',sum)
f.close()

f = open('kvlb.txt', 'r')
y = f.readlines()

print('the number of lines =',len(y))
```

WAP to copy one text file data into another text file data

```python
f = open('vow.txt', 'r')
x = f.readlines()

f1 = open('num2.txt', 'w')
for i in x:
  f1.write(i)

f.close()
f1.close()
```

WAP to count number of vowels in the given file

```python
f = open('vow.txt', 'w')
f.write('I am in class\nReading files\nWorking on python')
f.close()

f= open('vow.txt', 'r')
x = f.readlines()

L=['A','a', 'E','e','I','i','o','u','O','U']

count = 0
for i in x:
  for j in i:
    if j in L:
      count +=1

print(count)
```
