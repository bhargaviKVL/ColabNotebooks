pyplot is a sub module of matplotlib- visualising data

 pyplot helps us draw graph




plot(X- axis values, Y- axis values, color, linestyle, line width, marker, label)

Except Y- axis values everything else is optional

*   x- axis and y- axis values are given as lists
*   character assigned to color
*   ^ triangles
*   line width default = 1
*   markers - o, ^
*   Label = 'A'









```python
import matplotlib.pyplot as mpl

x = [1,2,3,4,5]
mpl.xlabel('marks1')

y = [10,20,15,16,35]
mpl.ylabel('marks2')

mpl.title('student details')
mpl.plot(x, y, 'r--', label ='A')#r-- will take color as r and linestyle as --

x1= [1,2,3,4,5]
y1= [18,25,30,32,35]

mpl.plot(x1,y1,'go', label = 'B')

mpl.legend() # gives details of each line in a small box above

mpl.grid()# provides grid lines
mpl.show()
```

```python
x =[1,2,3,4,5]
y =[5,6,7,8,9]
mpl.plot(x,y,'ro', label ='student1')
mpl.xlabel('marks')
mpl.ylabel('attendance')
mpl.title('details')

mpl.xlim(0,80)# limits the graph bw 0 to 80
mpl.xticks(x1, labels = ['a', 'b' ,'c' ,'d', 'e'])# gives labels to the x points


x1= [16,35,6,78,9]
y1 = [23,4,5,6,7]
mpl.plot(x1,y1,'b--', label ='student2', linewidth = 2, marker = 'o')
mpl.axis([1,80,0,60])# instead of x lima nd y lim axis takes xmin,xmax,ymin,ymax

x2 = [2,6,7,8,15]
y2 = [64,23,45,67,1]

mpl.plot(x2,y2,'g^', label = 'student3')
mpl.grid()
mpl.legend()
```

```python
mpl.subplot(2,1,1)

x =[1,2,3,4,5]
y =[5,6,7,8,9]
mpl.plot(x, y)
mpl.title('data')


mpl.subplot(2,1,2)

x2 = [2,6,7,8,15]
y2 = [64,23,45,67,1]
mpl.plot(x2,y2)
mpl.title('data1')
```

draw a graph to find number of students in each branch

```python
import pandas as pd
data = pd.read_excel('bgavi.xlsx', 'Sheet1')

grouped = data.groupby('Branch')
for branch, branch_df in grouped:
  print(branch)
  print(branch_df)

data.loc[data['Branch']=='aero', 'SNO'].count()
```

```python
x = ['cse', 'eece', 'mech', 'aero', 'ds', 'bs', 'aiml']
aero1 = data.loc[data['Branch']=='aero', 'SNO'].count()
cse1 = data.loc[data['Branch']=='cse', 'SNO'].count()
eece1 = data.loc[data['Branch']=='eece', 'SNO'].count()
mech1 = data.loc[data['Branch']=='mech', 'SNO'].count()
ds1 = data.loc[data['Branch']=='ds', 'SNO'].count()
bs1 = data.loc[data['Branch']=='bs', 'SNO'].count()
aiml1 = data.loc[data['Branch']=='aiml', 'SNO'].count()

y = [aero1,cse1,eece1,mech1,ds1,bs1,aiml1]
mpl.bar(x ,y, color = 'g', label = 'Students')
mpl.xlabel('branch')
mpl.ylabel('number of students')
mpl.legend()
```
