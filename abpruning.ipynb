```python
import time

board = [['.' for _ in range(3)] for _ in range(3)]
player_turn = 'X'

def is_end():
    for i in range(3):
        if board[i][0]==board[i][1]==board[i][2]!='.': return board[i][0]
        if board[0][i]==board[1][i]==board[2][i]!='.': return board[0][i]
    if board[0][0]==board[1][1]==board[2][2]!='.': return board[0][0]
    if board[0][2]==board[1][1]==board[2][0]!='.': return board[0][2]
    return '.' if all(board[i][j]!='.' for i in range(3) for j in range(3)) else None

def max_ab(alpha,beta):
    end=is_end()
    if end=='X': return -1,None
    if end=='O': return 1,None
    if end=='.': return 0,None
    maxv, move=-2,None
    for i in range(3):
        for j in range(3):
            if board[i][j]=='.':
                board[i][j]='O'
                score,_=min_ab(alpha,beta)
                board[i][j]='.'
                if score>maxv: maxv,move=score,(i,j)
                if maxv>=beta: return maxv,move
                alpha=max(alpha,maxv)
    return maxv,move

def min_ab(alpha,beta):
    end=is_end()
    if end=='X': return -1,None
    if end=='O': return 1,None
    if end=='.': return 0,None
    minv, move=2,None
    for i in range(3):
        for j in range(3):
            if board[i][j]=='.':
                board[i][j]='X'
                score,_=max_ab(alpha,beta)
                board[i][j]='.'
                if score<minv: minv,move=score,(i,j)
                if minv<=alpha: return minv,move
                beta=min(beta,minv)
    return minv,move

def print_board():
    for row in board: print(' '.join(row))
    print()

while True:
    print_board()
    if (res:=is_end())!=None:
        print("Winner:", res if res!='.' else "Tie")
        break

    # Human turn
    while True:
        start=time.time()
        _,rec_move=min_ab(-2,2)
        end=time.time()
        print("Evaluation time: {:.7f}s".format(end-start))
        print("Recommended move: X={}, Y={}".format(*rec_move))
        x,y=map(int,input("Insert your move X Y: ").split())
        if board[x][y]=='.':
            board[x][y]='X'
            break
        else:
            print("Invalid move! Try again.")

    # AI turn
    _,move=max_ab(-2,2)
    if move!=None:
        board[move[0]][move[1]]='O'
```
