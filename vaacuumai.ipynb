```python
# Agent function table
agent_table = {
    ('Clean', 'A'): 'MoveRight',
    ('Clean', 'B'): 'MoveLeft',
    ('Dirty', 'A'): 'Suck',
    ('Dirty', 'B'): 'Suck',
}

# Vacuum cleaner class
class VacuumCleaner:
    def __init__(self, location='A', status='Clean'):
        self.location = location
        self.status = status

    def percept(self):
        return self.status

    def act(self, action):
        if action == 'MoveRight':
            self.location = 'B'
        elif action == 'MoveLeft':
            self.location = 'A'
        elif action == 'Suck':
            self.status = 'Clean'

# Table-driven agent function
def table_driven_agent(percept):
    return agent_table.get(percept, 'NoOp')

# Main simulation loop with user input
if __name__ == "__main__":
    vacuum = VacuumCleaner()

    for _ in range(5):  # Run for 5 time steps
        location = input("Enter vacuum location (A or B): ").strip().upper()
        status = input("Enter status at location (Clean or Dirty): ").strip().capitalize()

        if location not in ['A', 'B'] or status not in ['Clean', 'Dirty']:
            print("Invalid input. Try again.\n")
            continue

        vacuum.location = location
        vacuum.status = status

        current_percept = vacuum.percept()
        action = table_driven_agent((current_percept, vacuum.location))
        print(f"Percept: ({current_percept}, {vacuum.location}), Action: {action}")

        if action != 'NoOp':
            vacuum.act(action)

        print(f"New Location: {vacuum.location}, New Status: {vacuum.status}\n")

```
