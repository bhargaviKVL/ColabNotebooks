PANDAS- module that helps us analyse large data stored in a tabular form


Data frames- key component of pandas (like file handler in file)

*  data frames can be cretaed from a excell file, CSV(comma separated value) file, dictionary, list of tuples
*   data in the form of a table



```python
import pandas as pd
d= {'id':[1,2,3,4], 'name':['b','h', 'r','a'], 'marks':[10,20,30,40]}
df = pd.DataFrame(d)
df
#df.name or df['name'] gives particular column data
#df[['name','marks']] gives list of multiple columns' data
```

max(), min(), sum(), mean(), std() are fns that ar available on integer data

```python
df['marks'].max()
df['marks'].mean()
```

shape gives order of table(matrix)

```python
df.shape
```

```python
df.mean()
# performs on only columns with integer data types
```
